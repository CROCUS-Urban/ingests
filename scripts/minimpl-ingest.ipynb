# Ingest Mini-MPL Data Using ACT

```python
import act
import glob
from datetime import datetime, timedelta
from distributed import Client
```

```python
client = Client()
client
```

```python
start_date = datetime(2023, 6, 13)
for i in range(80):
    this_date = start_date + timedelta(days=i)
    path_to_files = this_date.strftime("../../../../../crocus-data/mini-mpl/%Y%m%d*")
    files = sorted(glob.glob(path_to_files))

    if len(files) > 0:
        try:
            ds = act.io.mpl.read_sigma_mplv5(files,
                                             afterpulse='../../../../../crocus-data/mini-mpl/config/MMPL5037_Afterpulse_202303221000.bin',
                                             dead_time='../../../../../crocus-data/mini-mpl/config/MMPL5037_SPCM30891_Deadtime10.bin',
                                             overlap='../../../../../crocus-data/mini-mpl/config/MMPL5037_Overlap_202303281300.bin'
                                            ).compute()
            ds.time_utc.encoding.pop("_FillValue")
            ds.to_netcdf(this_date.strftime("crocus-neiu-minimpl-a1-%Y%m%d-%H%M%S.nc"), mode='w')
        except Exception as e:
            ds = None
            print(e)
    
        del ds
```
